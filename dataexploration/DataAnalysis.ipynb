```python
import pandas as pd
```

```python
## Read the Data
df=pd.read_csv('../data/nbl_eval.csv')
```

## Top 10 Expensive Companies:
Calculate the linehaulcost per mile and fuelcost per mile, summation of these two variables gives the total cost for each truckingcompanyid

```python
a=df
a.loc[:,'lhcpermile']=(a.linehaulcost/a.distanceinmiles).copy()
a.loc[:,'fuelpermile']=(a.fuelcost/a.distanceinmiles).copy()
a.loc[:,'tot_cost']=(a['lhcpermile']+a['fuelpermile']).copy()
```

```python
a.head()
```

```python
#idx1=a[['truckingcompanyid','tot_cost']].groupby('truckingcompanyid').sum().sort_values(by='tot_cost',ascending=False)[:10].index
```

```python
a_mean=a.groupby('truckingcompanyid')['tot_cost'].mean()

idx=a_mean.sort_values(ascending=False)[:10].index
a_mean.sort_values(ascending=False)[:10]

df[df.truckingcompanyid.isin(idx)].sort_values(by='tot_cost',ascending=False)
```

```python
df_expensive_companies=df[df.truckingcompanyid.isin(idx)].sort_values(by='truckingcompanyid')
```

```python
print("Top 10 most expensive trucking comapnies\n")
print(list(idx))
```

```python
df_expensive_companies.to_csv('../data/top_10_most_expensv_cmpy.csv',index=False)
```

## Top 10 most source-destinated companies

```python
## Calculate the most source-destination routes
b=df
b['agg']=1
sd=b.groupby(['source','destination'])['agg'].sum()
sd_df=pd.DataFrame(sd).reset_index()
sd_df.sort_values(by='agg',ascending=False).head()
```
