```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/nbl_eval.csv')
```

```python
df['truckingcompanyid']=df['truckingcompanyid'].astype(str)
df['productid']=df['productid'].astype(str)
```

```python
sns.pairplot(df.select_dtypes(exclude='object'))
```

```python
plt.figure(figsize=[10,6])
sns.boxplot(data=df, x='productid',y='linehaulcost')
plt.ylim([0,6000])
```

```python
df.source=df.source.astype('category').cat.codes
df.destination=df.destination.astype('category').cat.codes
```

```python
plt.figure(figsize=[10,6])
sns.boxplot(data=df, x='source',y='linehaulcost')
plt.ylim([0,2000])
```

```python
plt.figure(figsize=[10,6])
sns.boxplot(data=df, x='destination',y='linehaulcost')
plt.ylim([0,2000])
```

```python
sns.catplot(x='destination',y='linehaulcost',data=df,height=6, aspect=2)
plt.ylim([0,15000])
plt.xlim([0,51])
```

```python
sns.catplot(x='truckingcompanyid',y='linehaulcost',data=df,height=6, aspect=3)
```

```python
plt.figure(figsize=[10,6])
sns.catplot(data=df, x='destination',y='linehaulcost')
plt.show()
```

```python
sns.pairplot(df[['source','destination','linehaulcost']])
```

```python
sns.barplot(df.groupby('source')['truckingcompanyid'].count().sort_values(ascending=False))
```

```python
plt.figure(figsize=[15,5])
sns.barplot(x=df['destination'].sort_values(),y=df['linehaulcost'])
plt.xticks(rotation=40,fontsize=8)
plt.show()
```

```python
df=df.sample(5000)
plt.figure(figsize=[15,5])
sns.barplot(x=df['destination'].sort_values(),y=df['linehaulcost'])
plt.xticks(rotation=40,fontsize=8)
plt.show()
```
